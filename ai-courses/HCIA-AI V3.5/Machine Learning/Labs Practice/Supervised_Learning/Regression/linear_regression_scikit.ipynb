# Linear Regression

This notebook demonstrates how to fit a straight line to data in order to predict one variable from another.

**Usage:** to estimate unknown values based on existing data.

**In the code below, it is used to model the relationship between house area and price.**

**import the necessary libraries**

```python
from sklearn.linear_model import LinearRegression # used for creating the linear regression model
import matplotlib.pyplot as plt # used for drawing the graph
import numpy as np # used for numerical operations
```

**Prepare the data**

Assign the sample area and price data to variables for use in modeling.

```python
x = np.array([121, 125, 131, 141, 152, 161]).reshape(-1,1) # x targets the area of the house 
                                            # Reshape is used to convert the array into a 2D array with one column
                                            # required for the model to understand the input shape
y = np.array([517, 496, 405, 425, 350, 300]) # y targets the price of the house
plt.scatter(x,y) # draws the data points on the graph
plt.xlabel("area") # X axis indicates the area.
plt.ylabel("price") # Y axis indicates the price

plt.show() # Show the graph with the data points
```

**Visualize the data**

Create a scatter plot to see the relationship between area and price.

```python
# Train the model
model = LinearRegression() # Create an instance of the LinearRegression model
model.fit(x, y) # Fit the model to the data

# Visualize the model
y_pred = model.predict(x) # Predict the prices direction
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')  # draw the regression line
# The line is drawn using the predicted values
plt.xlabel("area")
plt.ylabel("price")
plt.title("Linear Regression: Area vs Price")
plt.show()
```

**Train and visualize the model**

Fit a linear regression model to the data and plot the regression line to see how well it fits the data.

```python
plt.scatter(x, y)
plt.xlabel("area") 
plt.ylabel("price") 
plt.plot([x[0][0], x[-1][0]], # draw the regression line
        [model.coef_[0]*x[0][0] + model.intercept_, # Calculate the y value for the first x
         model.coef_[0]*x[-1][0] + model.intercept_]) # Calculate the y value for the last x
# the line is drawn using 1st and last x values which is quicker than using the predicted values
```

```python
testX = np.array([[130]]) # A test sample with an area of 130
model.predict(testX) # Predict the price for the test sample
```
