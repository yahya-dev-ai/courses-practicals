# K-means Clustering

This notebook demonstrates how to use the K-means algorithm to group data into clusters based on feature similarity.

**Usage:** to identify natural groupings or patterns in data by assigning samples to the nearest cluster center.

**Import the necessary libraries**

```python
from sklearn.datasets import make_blobs # used to generate data for clustering
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans # used to perform K-means clustering
```

**Generate data**

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0) # Generate data with 4 clusters
```

** Display the dimension information of the generated dataset**

```python
print("Shape of X:", X.shape) # show the number of samples and features
print("Shape of y:", y.shape) # show the number of labels
```

**draw scatter graphs**

```python
plt.scatter(X[:, 0], X[:, 1], s=30, cmap='viridis') # Plot the generated data points
plt.title("Generated Data Points") # Set the title of the plot
plt.xlabel("Feature 1") # Label for the x-axis
plt.ylabel("Feature 2") # Label for the y-axis
```

**draw scatter graph in deffrent colors based on labels**

```python
plt.scatter(X[:, 0], # x-coordinates of the data points
             X[:, 1],# y-coordinates of the data points
               c=y,  # colors of the points based on their labels
                 s=30,# size of the points
                   cmap='viridis') # colormap used for coloring the points

plt.title("Data Points Colored by Labels") # Set the title of the plot
plt.xlabel("Feature 1") # Label for the x-axis
plt.ylabel("Feature 2") # Label for the y-axis
plt.show() # Display the plot  
```

**Perform k-means clustering**

```python
kmeans = KMeans(n_clusters=4) # Create a KMeans instance with 4 clusters
kmeans.fit(X) # Fit the model to the data       
```
